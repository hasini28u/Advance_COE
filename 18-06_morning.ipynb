```python
import pandas as pd
df = pd.read_pickle('/content/sample_data/2_CarPricesData.pkl')
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df[['Age','KM','Weight','HP','CC']]= sc.fit_transform(df[['Age','KM','Weight','HP','CC']])
df.head()
```

```python
X = df[['Age',	'KM',	'Weight',	'HP',	'MetColor',	'CC',	'Doors'	]]
y = df[['Price']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python

from keras import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping,ModelCheckpoint
model = Sequential()
model.add(Dense(units=7,input_dim=7,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=34,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=60,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=15,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=4,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='he_uniform'))
```

```python
model.compile(loss='mean_square_error',optimizer='adam',metrics=['mse'])
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience = 5,
    restore_best_weights=True,
    verbose=1
)
```

```python
checkpoint = ModelCheckpoint(
    'best_model.keras', #.keras is better than h5
    monitor='val_loss',
    verbose=1,
    save_best_only=True
)
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mse'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=20,verbose=1,callbacks=[early_stop,checkpoint],validation_split=0.5)
```
