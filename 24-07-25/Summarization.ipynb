```python
from transformers import BartTokenizer, BartForConditionalGeneration
import torch
#step1 import model
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
```

```python
text="""Hyd, short for Hyderabad, is the capital city of the Indian state of Telangana and also serves as the de jure capital of Andhra Pradesh. Known as the "City of Pearls," Hyderabad is famous for its rich history, vibrant culture, and thriving IT industry. The city boasts iconic landmarks like Charminar, Golconda Fort, and the Qutb Shahi Tombs, alongside a bustling food scene featuring local delicacies such as Hyderabadi biryani. Hyderabad is a major hub for technology companies, biotechnology, and pharmaceuticals, making it a prominent center for innovation and business in India."""
```

```python
inputs = tokenizer([text], max_length=1024, return_tensors='pt')
```

```python
summary_ids=model.generate(inputs['input_ids'],num_beams=4,max_length=100,early_stopping=True)
```
