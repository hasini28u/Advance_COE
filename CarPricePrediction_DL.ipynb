```python
import pandas as pd

# Load your dataset (assuming it's in a variable called df)
# df = pd.read_csv('your_file.csv')  # if needed

# Columns to check for outliers
num_cols = ['Age', 'KM', 'Weight', 'HP', 'CC', 'Price']

# Remove outliers using IQR
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Outliers removed. New shape:", df.shape)
```

```python
#since the values are apart from one another we need to scale them to a common scale
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df[['Age','KM','Weight','HP','CC']]= sc.fit_transform(df[['Age','KM','Weight','HP','CC']])
df.head()
```

```python
X= df.iloc[:,:-1]
y=df['Price']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from keras import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(units=7,input_dim=7,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=14,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=50,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=22,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='he_uniform'))
```

```python
model.summary()
```

```python

model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mse'])
model.fit(  X_train,y_train,epochs=50,batch_size=25,verbose=1,validation_split=0.5)
```
