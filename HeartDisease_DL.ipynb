```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/heart(in).csv')
```

```python
df
```

```python
df['Sex'] = df['Sex'].map({'M':0,'F':1})
df
```

```python
df['ChestPainType'].value_counts()
```

```python
df['ChestPainType']= df['ChestPainType'].map({'ATA':0,'NAP':1,'ASY':2,'TA':3})
df
```

```python
df['RestingECG'].value_counts()
```

```python
df['RestingECG'] = df['RestingECG'].map({'Normal':0,'ST':1,'LVH':2})
df
```

```python
df['ExerciseAngina'].value_counts()
```

```python
df['ExerciseAngina'] = df['ExerciseAngina'].map({'Y':0,'N':1})
df
```

```python
df['ST_Slope'].value_counts()
```

```python
df['ST_Slope'] = df['ST_Slope'].map({'Up':0,'Flat':1,'Down':2})
df
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df[['Age',	'Sex',	'ChestPainType','RestingBP','Cholesterol','FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope'	]]= sc.fit_transform(df[['Age',	'Sex',	'ChestPainType',	'RestingBP',	'Cholesterol',	'FastingBS',	'RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope'	]])
df.head()
```

```python
X = df[['Age',	'Sex',	'ChestPainType',	'RestingBP',	'Cholesterol',	'FastingBS',	'RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope'	]]
y = df[['HeartDisease']]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python

from keras import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping,ModelCheckpoint
model = Sequential()
model.add(Dense(units=7,input_dim=11,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=34,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=60,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=15,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=4,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='he_uniform'))
```

```python
model.compile(loss='mean_square_error',optimizer='adam',metrics=['mse'])
```

```python
early_stop = EarlyStopping(
    monitor='val_precision',
    patience = 5,
    restore_best_weights=True,
    verbose=1
)
```

```python
checkpoint = ModelCheckpoint(
    'best_model.keras', #.keras is better than h5
    monitor='val_precision',
    verbose=1,
    save_best_only=True
)
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['precision'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=20,verbose=1,callbacks=[early_stop,checkpoint],validation_split=0.5)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
```
