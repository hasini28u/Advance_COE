```python
category_symptom_ct = data.groupby('Category Id')['user_symptom'].agg(lambda x: x.value_counts().index[0])
category_symptom_ct_freq = data.groupby('Category Id')['user_symptom'].value_counts().groupby('Category Id').max() / data.groupby('Category Id')['user_symptom'].count()

print(pd.DataFrame({
    'Dominant Symptom': category_symptom_ct,
    'Dominant Frequency %': (category_symptom_ct_freq * 100).round(2)
}))
```

```python
group_symptom_ct = data.groupby('Support_group')['user_symptom'].agg(lambda x: x.value_counts().index[0])
group_symptom_ct_freq = data.groupby('Support_group')['user_symptom'].value_counts().groupby('Support_group').max() / data.groupby('Support_group')['user_symptom'].count()

print(pd.DataFrame({
    'Dominant Symptom': group_symptom_ct,
    'Dominant Frequency %': (group_symptom_ct_freq * 100).round(2)
}))
```

splitting data into train and test

```python
from sklearn.model_selection import train_test_split
X = df_model_clean.drop(columns=['impact'])
y = df_model_clean['impact']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

fitting into various models

```python
# Logistic Regression
lr = LogisticRegression(max_iter=500)
lr.fit(X_train, y_train)
```

```python
# Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

```python
# XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)
```

evaluating models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

models = {'Logistic Regression': lr, 'Decision Tree': dt, 'Random Forest': rf, 'XGBoost': xgb}

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"--- {name} ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
    print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
    print("F1 Score (macro):", f1_score(y_test, y_pred, average='macro'))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
