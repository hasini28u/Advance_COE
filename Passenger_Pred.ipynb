```python
import pandas as pd
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])

# Create separate columns for date and time
df['pickup_date'] = df['pickup'].dt.date
df['pickup_time'] = df['pickup'].dt.time

df['dropoff_date'] = df['dropoff'].dt.date
df['dropoff_time'] = df['dropoff'].dt.time
```

```python
df.drop(['pickup','dropoff'],axis=1,inplace=True)
```

```python
print(df['payment'].value_counts())
print(df['color'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ['color', 'payment']
df_encoded = df.copy()
le = LabelEncoder()
for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df_encoded[['distance','fare','tip','tolls']]= sc.fit_transform(df_encoded[['distance','fare','tip','tolls']])
df.head()
```

```python
df_encoded = df_encoded.select_dtypes(include=['number'])
```

```python
df_encoded
```

```python
X = df_encoded.iloc[:, :8]
y=df_encoded['total']
print(X.shape)
print(y.shape)
```

```python
X = X.drop('total', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from keras import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(units=7,input_dim=7,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=14,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=50,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=22,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='he_uniform'))
```

```python

model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mse'])
model.fit(  X_train,y_train,epochs=50,batch_size=25,verbose=1,validation_split=0.5)
```
